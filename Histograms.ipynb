# Histograms:

To construct a histogram, the first step is to "bin" the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval. The bins are usually specified as consecutive, non-overlapping intervals of a variable. 

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
dataset1 = np.random.randn(100) # 100 random values
```

```python
# Let's plot the histogram for the above dataset
plt.hist(dataset1)
plt.show()
```

If you notice the above chart, you will see that there are 10 bins or bars, thats because hist by default uses 10 bins

```python
dataset2 = np.random.randn(80)

plt.hist(dataset2, color="indianred")
plt.show()
```

Let's plot two histograms together:

```python
plt.hist(dataset1, normed=True, color="blue", bins=20, alpha=0.5)
plt.hist(dataset2, normed=True, color="indianred", bins=20, alpha=0.5)

plt.show()
```

In the above histogram plots:

1. <i>normed = True </i>, Since both the dataset had different number of datasets we use this to normalize the two dataset.
2. <i>bins = 20</i>, By default the number of bins is 10, we can increase or decrease the number of bins to use using this.
3. <i>alpha = 0.5</i>, This is used to set transparency, the value is between 0 to 1
